# 4) Handle outliers 

For the variable â€œFareâ€, replace outlier values with the boundary values


*   If value < (Q1 - 1.5IQR), relace with (Q1 - 1.5IQR)
*   If value > (Q3 + 1.5IQR), relace with (Q3 + 1.5IQR)

<font color='blue'>Q4) What is the mean of â€œFareâ€ after replacing the outliers (round 2 decimal points)?</font>
 
ans: 24.04


```python
q25 = df["Fare"].quantile(0.25)
q75 = df["Fare"].quantile(0.75)
iqr = q75 - q25

min = q25 - (iqr*1.5)
max = q75 + (iqr*1.5)
print('Min: ', min)
print('Max: ', max)
```

```python
df["Fare"] = df["Fare"].where(df["Fare"] >= min, min)
df["Fare"] = df["Fare"].where(df["Fare"] <= max, max)
```

```python
df['Fare'].mean()
```

# 5) Impute missing value



*   Show the number of missing values in each variable
*   Impute missing values with mean & mode
*   Show the number of missing values again after missing value imputation

<font color='blue'>Q5) Which variable has the largest number of missing values?</font>

ans: Age (172)



```python
has_nan_list = []
for var in df.columns.tolist():
    missing_values = df[var].isna().sum()
    print(var, missing_values)
    if missing_values > 0: has_nan_list.append(var)
has_nan_list
```

```python
for col in has_nan_list:
    if col in cate_cols:
        df[col] = df[col].fillna(df[col].mode().iloc[0])
    else: df[col] = df[col].fillna(df[col].mean())
```

```python
for var in df.columns.tolist():
    missing_values = df[var].isna().sum()
    print(var, missing_values)
```
