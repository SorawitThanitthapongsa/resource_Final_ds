## Compare "Sports" and "Comady", how many days that there are more total daily views of VDO in "Sports" category than in "Comady" category?

```python
import json

with open('data/US_category_id.json') as fd:
    cat = json.load(fd)
```

```python
cat_id = []
for d in cat['items']:
    cat_id.append((int(d['id']), d['snippet']['title']))
```

```python
cat_df = pd.DataFrame(cat_id, columns=['category_id', 'category'])
cat_df
```

```python
df_withcat = pd.merge(df, cat_df, on = 'category_id')
df_withcat
```

```python
df_comedy_bydate = df_withcat.loc[df_withcat['category'] == 'Comedy'].groupby('trending_date')
df_comedy_totalviews = df_comedy_bydate.views.sum()
df_comedy_totalviews
```

```python
df_sports_bydate = df_withcat.loc[df_withcat['category'] == 'Sports'].groupby('trending_date')
df_sports_totalviews = df_sports_bydate.views.sum()
df_sports_totalviews
```

```python
df_sports_totalviews[df_sports_totalviews > df_comedy_totalviews].count()
```
