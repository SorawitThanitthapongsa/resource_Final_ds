# Step 2. Drop rows where the target (label) variable is missing.

```python
df = df.dropna(subset=['label'])
df
```

# Step 3. Drop the following variables:

```python
drop_variable = ['id','gill-attachment', 'gill-spacing', 'gill-size','gill-color-rate','stalk-root', 'stalk-surface-above-ring', 'stalk-surface-below-ring', 'stalk-color-above-ring-rate','stalk-color-below-ring-rate','veil-color-rate','veil-type']
for i in drop_variable:
  df = df.drop(columns=[i])
df
```

```python
df.shape
```

# Step 5. Fill missing values by adding the mean for numeric variables and the mode for nominal variables.

```python
for i in df.columns:
  null_count = df[i].isnull().sum()
  print(i,"has null values = ", null_count)
```

```python
df.nunique()
```

```python
df.info()
```

```python
df.fillna(df.mean(), inplace=True)
df.fillna(df.mode().iloc[0], inplace=True)
```

# Step 6. Convert the label variable e (edible) to 1 and p (poisonous) to 0 and check the quantity. class0: class1

```python
df = df.replace(["p","e"],[0,1])
```

# Step 7. Convert the nominal variable to numeric using a dummy code with drop_first = True.

```python
df_dummies = pd.get_dummies(df[["cap-shape","cap-surface","bruises","odor","stalk-shape","ring-number","ring-type","spore-print-color","population","habitat"]], drop_first=True)
```

```python
df = pd.get_dummies(df, drop_first=True)
```

# Step 8. Split train/test with 20% test, stratify, and seed = 2020.

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
y = df.pop('label')
X = df
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=2020)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Step 9. Create a Random Forest with GridSearch on training data with 5 CV.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report,confusion_matrix
```

```python
model = GridSearchCV(estimator=RandomForestClassifier(),param_grid=dict(criterion=['gini','entropy'],max_depth=[2,3,6],min_samples_leaf=[2,5,10],
    n_estimators=[100,200], random_state=[2020]),scoring='f1_weighted',cv=5,n_jobs=-1
) 
```

```python
model.fit(X_train, y_train)
```

```python
print('Best param is')
print(model.best_params_)
```

```python
model = model.best_estimator_
```

```python
y_pred = model.predict(X_test)
```

```python
print(classification_report(y_test,y_pred,digits=4))
```

```python
print(confusion_matrix(y_test,y_pred,labels=[0,1]))
```
