## What is the average number of follows per user?

```python
total_follow = 0
all_user_follows = rd.keys('user:*:follows')
for userid in all_user_follows:
    total_follow += rd.scard(userid)
print(total_follow/all_user_id.shape[0])
```

## How many users follows between 5-10 users?

```python
num_count = 0
for userid in all_user_follows:
    num_follows = rd.scard(userid)
    if num_follows >= 5 and num_follows <= 10:
        num_count +=1
print(num_count)
```

## Which account has the most followers?

```python
max_followers = -1
account = -1
all_user_followedby = rd.keys('user:*:followed_by')
for userid in all_user_followedby:
    num_followers = rd.scard(userid)
    if num_followers > max_followers:
        account = userid
        max_followers = num_followers
print(account)
```
